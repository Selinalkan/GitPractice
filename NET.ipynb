```python
#imports
import nltk
from nltk.tokenize import word_tokenize
from nltk.tag import pos_tag
from nltk.chunk import ne_chunk
from nltk.chunk import conlltags2tree, tree2conlltags
from pprint import pprint
import spacy
from spacy import displacy
from collections import Counter
import en_core_web_sm
import pandas as pd
from typing import List, Tuple
```

```python
assert nltk.download('averaged_perceptron_tagger')
assert nltk.download('conll2000')
assert nltk.download("punkt")
assert nltk.download('maxent_ne_chunker')
assert nltk.download('words')
```

```python
#SpaCy entity
nlp = en_core_web_sm.load()
```

# **Interviewer with Context**

```python
#Importing ER_c_Transcript Dataset and dropping the Unnamed column
df = pd.read_csv("ER_c_transcript.csv").drop("Unnamed: 0", axis=1)
```

```python
#Sorting the data by "speaker_trustworthy" values
df_sorted1 = df.sort_values("speaker_trustworthy", ascending=False)
#Sorting the data by "speaker_trusted by conversational partner" values
df_sorted2 = df.sort_values("speaker_trusts conversational partner", ascending=False)
```

```python
#Joining the first 10 rows to create 1 word cloud per feature
df_sorted1_joined = " ".join(df_sorted1.ER_text[:10])
df_sorted2_joined = " ".join(df_sorted2.ER_text[:10])
```

**"speaker_trustworthy"**

```python
df1 = nlp(df_sorted1_joined)
pprint([(X.text, X.label_) for X in df1.ents])
```

```python
labels = [x.label_ for x in df1.ents]
Counter(labels)
```

```python
displacy.render(nlp(str(df1)), jupyter=True, style='ent')
```

**"speaker_trusts conversational partner"**

```python
df2 = nlp(df_sorted2_joined)
pprint([(X.text, X.label_) for X in df1.ents])
```

```python
labels = [x.label_ for x in df2.ents]
Counter(labels)
```

```python
displacy.render(nlp(str(df2)), jupyter=True, style='ent')
```
